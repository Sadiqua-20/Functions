### Q1. What is the difference between a function and a method in Python?

ANS 1.

In Python, the terms function and method are related but have distinct meanings based on their usage and context. Here's a comparison:

1. ***Function***

- A function is a block of reusable code that performs a specific task.

- It is defined using the **def** keyword and can exist independently, outside any class.

- It is called directly using its name and can take arguments and return a value.

- Example:

  ```
  def add(a, b):
      return a + b

  print(add(3, 5))  
  ```

  OUTPUT:

  ```
  8
  ```

2. ***Method***

- A method is essentially a function that is associated with an object or class.

- It is defined inside a class and operates on the object (instance) or class it belongs to.

- A method typically has at least one parameter, often named **self** for instance methods, to reference the object it operates on.

- Example:

  ```
  class Math:
      @staticmethod
      def add(a, b):
          return a + b

  print(Math.add(3, 5))
  ```

  OUTPUT:

  ```
  8
   ```

### Q2. Explain the concept of function arguments and parameters in Python.

ANS 2.

In Python, function arguments and parameters are terms used to describe the data passed to functions and how functions accept that data. Here's an explanation of these concepts:

1. **Parameters**

- Parameters are the variables defined in the function's declaration.

- They act as placeholders for the values (arguments) that will be provided when the function is called.

- A function can have one or more parameters, or none at all.

- Example:

  ```
  def greet(name):  # 'name' is a parameter
      return f"Hello, {name}!"
  ```
  

2. **Arguments**

- Arguments are the actual values or data passed to the function when it is called.

- The arguments are assigned to the corresponding parameters defined in the function.

- Example:

  ```
  greet("SADIQUA")   # "SADIQUA" is an argument
  ```



### Q3. What are the different ways to define and call a function in Python?

In Python, functions can be defined and called in several ways, depending on their type, purpose, and use case. Here's an overview:

1. **Regular Function**

- A regular function is defined using the **def** keyword, followed by the function name, parameters, and a block of code.

- Example:

  ```

  ```

  ```

  ```

- Functions can include default values for parameters.

- Example:

  ```

  ```
  Call:

  ```
  print(greet())          
  print(greet("SADIQUA"))  
  ```
  Output:

  ```
  Hello, Guest!
  Hello, SADIQUA!
  ```

3. **Function with Keyword Arguments**

- Functions can accept arguments explicitly using parameter names.

- Example:

  ```
  def introduce(name, age):
      return f"My name is {name} and I am {age} years old."
  ```
  Call:

  ```
  print(introduce(name="SADIQUA", age=24))  
  ```
  Output:

  ```
  My name is SADIQUA and I am 24 years old.
  ```

4. **Function with Variable-length Arguments**

- ` *args `: For an arbitrary number of positional arguments.

- ` **kwargs `: For an arbitrary number of keyword arguments.

- Example:
  ```
  def add(*numbers):
      return sum(numbers)

  def display_info(**info):
      return info
  ```
  Call:

  ```
  print(add(1, 2, 3))
  print(display_info(name="Alice", age=25))  
  ```
  Output:

  ```
  6
  {'name': 'Alice', 'age': 25}
  ```

5. **Lambda Function**

- A lambda function is a small, anonymous function defined using the lambda keyword. It is typically used for short operations.

- Example:

  ```
  square = lambda x: x * x
  ```
  Call:

  ```
  print(square(5))  
  ```
  Output:

  ```
  25
  ```

6. **Nested Function**

- A function defined inside another function.

- Example:

  ```
  def outer_function(name):
      def inner_function():
          return f"Hello, {name}!"
      return inner_function()
   ```
   Call:

   ```
   print(outer_function("SADIQUA"))  
   ```
   Output:

   ```
   Hello, SADIQUA!
   ```

7. **Recursive Function**

- A function that calls itself.

- Example:
  ```
  def factorial(n):
      if n == 1:
          return 1
      return n * factorial(n - 1)
  ```
  Call:

  ```
  print(factorial(5))

  Output:

  ```
  120
  ```

8. **Generator Function**

- Uses yield to return one value at a time, allowing iteration.

- Example:
  ```

  Call:

  ```
  for num in generate_numbers():
      print(num)

  Output:

  ```
  1
  2
  3
  ```

9. **Calling a Function via ` *args ` and ` **kwargs `**

- Pass arguments to a function dynamically using *args or **kwargs.

- Example:
  ```
  def greet(name, age):

  args = ("SADIQUA", 24)

### Q4. What is the purpose of the `return` statement in a Python function?

ANS 4.

The `return` statement in a Python function is used to send a value or result back to the caller of the function. It serves several purposes in making functions reusable and meaningful.

1. **Provide Output from a Function**:

### Q5. What are iterators in Python and how do they differ from iterables?

1. **ITERATORS**:

- An iterator is an object representing a stream of data. It produces items one at a time when you call the `next()` function on it.

- It implements the `iter()` method, which returns itself.

  ```

  ----> 6 print(next(l))  # Raises StopIteration

  StopIteration:
  ```

2. **ITERABLES**:

- An iterable is any Python object capable of returning its members one at a time. It is an object that can be "iterated over," meaning it can produce an iterator.

- Examples include lists, tuples, strings, dictionaries, sets, and more.

- An iterable has a special method called `iter()` that returns an iterator.

- Example:

  my_list = [1, 2, 3]       # List is an iterable

### Q6.  Explain the concept of generators in Python and how they are defined.

ANS 6.

1. **CONCEPT OF GENERATORS**:

- Generators are special type of functions in Python that allow us to yield values one at a time instead of returning them all at once. They are used to return an iterator object.

- **Lazy Evaluation**: Generators produce items on demand. Instead of storing all values in memory, they generate values as needed.
This makes them memory-efficient for large datasets or infinite sequences.

- **` yield ` Instead of ` return `**: Generators use the yield keyword to produce values one at a time.
When a generator function is called, it does not execute immediately; instead, it returns a generator object.

2. **DEFINING A GENERATOR**:

### Q7.  What are the advantages of using generators over regular functions?

ANS 7.

**Advantages of Using Generators Over Regular Functions**:

1. **Memory Efficiency**:
- Generators use lazy evaluation, meaning they produce items one at a time and do not store the entire dataset in memory.

- In contrast, regular functions typically return entire datasets, consuming more memory.

- Example:

  ```
  def generate_numbers(limit):
      for i in range(limit):
          yield i       # Generator

  # Generator doesn't store all numbers at once
  gen = generate_numbers(10**6)
  print(next(gen))      # Output: 0
  print(next(gen))      # Output: 1
  ```

  OUTPUT:

  ```
  0
  1
  ```

2. **Faster Execution for Large Iterations**:

- Generators don’t compute all results at once. They compute and yield values on demand, reducing initial computation time.
Regular functions may take longer if they process and return large datasets in one go.

- Example:
  ```
  # Generator (lazy evaluation)
  def squares_generator(n):
      for i in range(n):
          yield i * i

  # Regular function (precomputes all results)
  def squares_list(n):
      return [i * i for i in range(n)]

  # For large values of n, the generator will start producing values faster.
  ```

### Q9.  Explain the purpose and usage of the `map()` function in Python.

ANS 9.

1. **PURPOSE OF `map()` FUNCTION**:
- The `map()` function in Python is used to apply a given function to all items in an iterable (like a list, tuple, etc.) and return an iterator (map object) that yields the results of applying the function.
The main purpose of `map()` is to transform or modify each item of an iterable in a simple and concise manner, rather than using a loop.

2. **USAGE OF `map()` FUNCTION**:
- Manipulating Strings:

  ```
  words = ["hello", "world", "python"]

  # Use map to convert all words to uppercase
  uppercase_words = map(str.upper, words)

  # Convert map object to list
  print(list(uppercase_words))  
  ```
  OUTPUT:
  ```
  ['HELLO', 'WORLD', 'PYTHON']
  ```

- Applying a Function to Each Element in a List:  

  Let's say we want to double each number in a list using `map()`.

  ```
  numbers = [1, 2, 3, 4, 5]

  # Use map to apply a function (lambda) to each element in the list
  doubled = map(lambda x: x * 2, numbers)

  # Convert map object to list to display the results
  print(list(doubled))  
  ```

  OUTPUT:
  ```
  [2, 4, 6, 8, 10]
  ```

### Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

ANS 10.

1. **` map() ` Functions**:
- Definition: The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.

- Key Points:
  - Transformation: Used to transform each item in an iterable by applying the specified function.
- Syntax:
  ```
  map(function, iterable)
  ```
- Example:
  ```
  num = [1,2,3]
  square = map(lambda x: x**2, num)
  print(list(square))
  ```

  OUTPUT:
  ```
  [1, 4, 9]
  ```

2. **` reduce ` Functions**:
- Definition: The reduce function from the functools module applies a given function cumulatively to the items of a sequence, from left to right, to reduce the sequence to a single value.

- Key Points:

  - Cumulative Operation: It reduces an iterable to a single cumulative value by applying the function cumulatively.
- Syntax:
  ```
  from functools import reduce
  reduce(function, iterable)
  ```
- Example:
  ```
  from functools import reduce
  num = [1,2,3]
  sum = reduce(lambda x,y: x+y, num)

  print(sum)
  ```
   OUTPUT:

  ```
  6
  ```

3. **` filter ` Functions**:
- Definition: The filter function constructs an iterator from elements of an iterable for which a specified function returns True.

- Key Points:
  - Selection: Used for filtering elements from an iterable based on a condition defined by the function.
- Syntax:
  ```
  filter(function, iterable)
  ```

- Example:
  ```
  num = [1, 2, 3, 4, 5, 6, 7, 8]
  even_no = filter(lambda x: x%2 == 0, num)
  print(list(even_no))
  ```
  OUTPUT:

  ```
  [2, 4, 6, 8]
  ```

````python
def sum_of_even(LIST):
  s = 0
  for i in LIST:
    if i % 2 == 0:
      s = s + i

  return s
````

### Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
numbers = list(range(1,201))

def prime(num):
  if num <= 1:
    return False

  for i in range(2,num):
    if num % i == 0:
      return False
  return True

print("Prime numbers from 1 to 200 are: ")


for num in numbers:
  if prime(num):
    print(num, end = " ")
````

### Q5. . Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

````python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

n = 100
fib_iterator = FibonacciIterator(n)
for i, fib_num in enumerate(fib_iterator):
    print(f"Fibonacci number {i+1}: {fib_num}")
````

### Q7. Implement a generator function that reads a file line by line and yields each line as a string.



````python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
    print(line)
````

### Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

````python
def celsius_to_fahrenheit(celsius):
  return (celsius * (9/5)) + 32
````

````python
celsius = [0, 20, -40, 30, 100]
fahrenheit = list(map(celsius_to_fahrenheit, celsius))

print(" Temperatures in Fahrenheit: ", fahrenheit)
````

### Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, input_string))
````

````python
input_str = "Hello, how are you?"
result = remove_vowels(input_str)
print(f"Original string: {input_str}")
print(f"String without vowels: {result}")
````
